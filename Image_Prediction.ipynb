```python
import numpy as np
import keras
import matplotlib.pyplot as plt
```

```python
data = keras.datasets.fashion_mnist
```

```python
(train_images , train_labels), (test_images , test_labels) = data.load_data()
```

```python
class_names = ['T-shirt','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle-Boot']
```

```python
# Scaling our data b/w 0 and 1. The pixels for the image is spread b/w 0 and 255 therefore we are kind of doing 
#a max-Scaling.

train_images = train_images/255
test_images = test_images/255
```

```python
# Initializing The model
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),   ## Data is flattened to make it a list of 28x28 and remove all Complexity.
    keras.layers.Dense(128,activation='relu'),   ## Creating a hidden layer to increase the performance of the neurons.Relu-Rectified Linear unit.
    keras.layers.Dense(10,activation='softmax')  ## Softmax basically selects the neutron with highest probability.
])
```

```python
# Compiling and traning our model.
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

model.fit(train_images,train_labels,epochs=10)
```

```python
# Evaluating our Model
test_loss , test_acc = model.evaluate(test_images,test_labels)
```

```python
print('The model Accuracy is :',test_acc)
```

```python
print ('The model loss is: ',test_loss)
```

```python
prediction = model.predict(test_images)
```

```python
print(class_names[np.argmax(prediction[0])])
```

```python
# Printing the Actual and the Prediction of the images in test data along with the Images. 
for i in range(20):
    plt.grid(False)
    plt.imshow(test_images[i], cmap = plt.cm.binary)
    plt.xlabel("Actual : " + class_names[test_labels[i]])
    plt.title("Prediction : " + class_names[np.argmax(prediction[i])])
    plt.show()
```
